```python
quartil_2= df['Fare'].quantile(0.25)
quartil_3= df['Fare'].quantile(0.75)
```

```python
inter_quartil = quartil_3 - quartil_2
limite_superior= quartil_3 + (inter_quartil * 1.5)
limite_inferior= quartil_2 - (inter_quartil * 1.5)
```

```python
(limite_inferior,limite_superior)
```

```python
df.query("Fare > @limite_superior")
```

```python
std = df['Fare'].std()
limite_superior_std = df['Fare'].mean()+(2 * std)
limite_inferior_std = df['Fare'].mean()-(2 * std)
```
